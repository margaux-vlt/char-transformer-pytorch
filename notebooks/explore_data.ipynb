# Exploration of dataset

```python
import os
import requests

# Go one level up to repo root
os.makedirs("../data", exist_ok=True)

# Download Alice in Wonderland
url = "https://www.gutenberg.org/files/11/11-0.txt"
text = requests.get(url).text

# Optional: remove Gutenberg header/footer
start = text.find("*** START OF THIS PROJECT GUTENBERG EBOOK")
end = text.find("*** END OF THIS PROJECT GUTENBERG EBOOK")
if start != -1 and end != -1:
    text = text[start:end]

# Save file in top-level data folder
with open("../data/alice_in_wonderland.txt", "w", encoding="utf-8") as f:
    f.write(text)

print("Length of text:", len(text))
print(text[:500])
```
